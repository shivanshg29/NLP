```python
import pandas as pd
data=pd.read_pickle('./pickle/data_clean.pkl')
data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(stop_words='english')
data_dtm=cv.fit_transform(data.transcript)
data_dtm=pd.DataFrame(data_dtm.toarray(),columns=cv.get_feature_names_out())
data_dtm.index=data.index
data_dtm
```

```python
data_dtm=data_dtm.transpose()
data_dtm
```

```python
top_dict={}
for column in data_dtm.columns:
    top=data_dtm[column].sort_values(ascending=False).head(30)
    top_dict[column]=list(zip(top.index,top.values))

top_dict
```

```python
from collections import Counter
```

```python
words=[]
for comedian in data_dtm.columns:
    top=[word for (word,cnt) in top_dict[comedian]]
    for t in top:
        words.append(t)

words
```

```python
Counter(words).most_common()
```
